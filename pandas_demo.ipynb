## Cleaning Data

```python
people = {
    'first': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'email': ['j-m@email.com','m-j@email.com',np.nan, None, 'a-c@email.com'],
    'job': [None, 'Programmer', 'Scientist', 'Engineer', 'Doctor'],
    }

people_df = pd.DataFrame(people)
people_df
```

```python
# remove any rwo have NaN/None value in it
people_df.dropna()
# this is the same as: people_df.dropna(axis='index' , how='any')
```

```python
# remove any column have NaN/None value in it
people_df.dropna(axis='columns')
```

```python
people2 = {
    'first': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy', None],
    'last': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', np.nan],
    'email': ['j-m@email.com','m-j@email.com',np.nan, None, 'a-c@email.com', None],
    'job': [None, 'Programmer', 'Scientist', 'Engineer', 'Doctor', np.nan],
    }

people2_df= pd.DataFrame(people2)
people2_df
```

```python
# delete the rows that are entirety NaN/None
people2_df.dropna(axis='index' , how='all')
```

```python
# remove any column have NaN/None value in it
people2_df.dropna(axis='columns')
```

```python
# remove any row have NaN/None value in job column
people2_df.dropna(axis='index', subset=['job'])
```

```python
# remove any row have NaN/None value in both job column and email column
people2_df.dropna(axis='index', how='all' , subset=['job', 'email'])
```

```python
# remove any row have NaN/None value in job column or email column
people2_df.dropna(axis='index', how='any' , subset=['job', 'email'])
```
