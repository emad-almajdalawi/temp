## Filtering Data

```python
# or
jordan = df[df['Country'] == 'Jordan']
jordan.head(2)
```

```python
# employment status of Jordanians
df.loc[df['Country'] == 'Jordan', 'Employment'].value_counts()
```

```python
filt = (df['Country'] == 'Jordan') & (df['Employment'] == 'Employed full-time')
df.loc[filt, 'MainBranch']
```

```python
# ~ is the negation operator
filt = ((df['Country'] == 'Jordan') | (df['Country'] == 'Qatar')) & ~(df['Employment'] == 'Employed full-time')
df.loc[filt, 'MainBranch'].value_counts()
```

```python
# or
countries = ['Jordan', 'Qatar']
filt = (df['Country'].isin(countries)) & ~(df['Employment'] == 'Employed full-time')
df.loc[filt, 'MainBranch'].value_counts()
```

```python
df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
# or
#  filt = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
# df.loc[filt]
```

```python
filt = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
df.loc[filt, 'LanguageHaveWorkedWith']
```
